```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
```

```python
breast_cancer_data= sklearn.datasets.load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_dataframe = pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
breast_cancer_dataframe
```

```python
breast_cancer_dataframe["label"]=breast_cancer_data.target
breast_cancer_dataframe
```

```python
x=breast_cancer_dataframe.drop(columns="label",axis=1).values
y=breast_cancer_dataframe["label"].values
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
data_lg=LogisticRegression()
data_lg.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

#accuracy of training data
x_train_prediction =data_lg.predict(x_train)
training_data_accuracy= accuracy_score(y_train,x_train_prediction)
print("Training Data Accuracy : " ,training_data_accuracy)

#accuracy of test data
y_test_prediction =data_lg.predict(x_test)
testing_data_accuracy= accuracy_score(y_test,y_test_prediction)
print("Testing Data Accuracy : " ,testing_data_accuracy)
```
